### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate the x values from 0 to 10
x = np.arange(0, 11)

# Calculate the y values as the square of x
y = x**2

# Create the line plot
plt.plot(x, y)

# Add a title to the plot
plt.title("Line Plot of y = x^2")

# Add labels to the x-axis and y-axis
plt.xlabel("X-Axis Label (x)")
plt.ylabel("Y-Axis Label (x^2)")

# Add a grid to the plot
plt.grid(True)

# Display the plot
plt.show()
```
