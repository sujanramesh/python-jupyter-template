### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Line Chart in the first subplot
# Generate the x values from 0 to 10
x1 = np.arange(0, 11)
# Calculate the y values as the square of x
y1 = x1**2

# 2. Bar chart in the second subplot
# Data for the bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
# Define colors for each bar
colors = ['red', 'green', 'blue', 'purple']

# 3. Scatter plot in the third subplot
# Data for the scatter plot
x3 = [1, 2, 3, 4, 5]
y3 = [2, 4, 6, 8, 10]

# 4. Pie chart in the fourth subplot
# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
plt.figure(figsize=(10, 8))  # Adjust figure size for better visualization

# Add the line chart to the first subplot (1, 1)
plt.subplot(2, 2, 1)  # (rows, columns, plot_number)
plt.plot(x1, y1)
plt.title("Line Plot")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.grid(True)

# Add the bar chart to the second subplot (1, 2)
plt.subplot(2, 2, 2)
plt.bar(categories, values, color=colors)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

# Add the scatter plot to the third subplot (2, 1)
plt.subplot(2, 2, 3)
plt.scatter(x3, y3)
plt.title('Scatter Plot')
plt.xlabel("x")
plt.ylabel("y")
for i in range(len(x3)):
    plt.annotate(f'({x3[i]}, {y3[i]})', (x3[i], y3[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Add the pie chart to the fourth subplot (2, 2)
plt.subplot(2, 2, 4)
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')

# Adjust the layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plots
plt.show()
```
